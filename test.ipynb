# Connect to API EndPoint

```python
import asyncio
import websockets
import json
import base64
import nest_asyncio

nest_asyncio.apply()

async def send_audio(websocket_url, audio_file_path, translation_preference=None):
    # Open the audio file and read its content as binary
    with open(audio_file_path, "rb") as audio_file:
        audio_data = audio_file.read()

    # Encode the audio data to base64
    base64_audio_string = base64.b64encode(audio_data).decode('utf-8')

    async with websockets.connect(websocket_url) as websocket:
        # Set translation preference if provided
        if translation_preference:
            await websocket.send(json.dumps({
                "translation_preference": translation_preference
            }))
            response = await websocket.recv()
            print(f"Translation preference response: {response}")
        
        await websocket.send(json.dumps({
            "audio_data": base64_audio_string
        }))
        
        response = await websocket.recv()
        print(f"Received response: {response}")

# Replace with your actual WebSocket URL
websocket_url = "ws://127.0.0.1:8000/ws"

# Path to the audio file you want to send (e.g., an MP3 or WAV file)
audio_file_path = "verse_3.ogg"

# Optionally set the translation preference (e.g., 'english', 'arabic', etc.)
translation_preference = "arabic"
```

```python
asyncio.get_event_loop().run_until_complete(
    send_audio(websocket_url, audio_file_path, translation_preference)
)
```
